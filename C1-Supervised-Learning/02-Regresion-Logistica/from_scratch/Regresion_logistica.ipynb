```python
import numpy as np

# Dataset de prueba (m=3 ejemplos, n=1 característica)
X = np.array([[1], [2], [3]])       # Entrada
y_real = np.array([[2], [4], [6]])  # Salida esperada

# Condiciones iniciales del sistema
W = np.array([[0.0]])
b = 0.0
alpha = 0.01
epocas = 100


def predecir(dataset: np.ndarray, pesos: np.ndarray, sesgo: float) -> np.ndarray:
  return np.dot(dataset, pesos) + sesgo

def calcular_error(predicciones :np.ndarray, reales :np.ndarray):
  assert predicciones.shape == reales.shape
  error_array = predicciones - reales
  return error_array

def calcular_costo(error: np.ndarray) -> float:
  # el error lo multiplicas por si mismo transpuesto
  # ese ultimo se multiplica por 1/2*error.shape[0]
  costo = (1/(2 * error.shape[0])) * (np.dot(error.T, error))
  return costo

def calculo_gradiente(dataset: np.ndarray, errores: np.ndarray):
  return (1/dataset.shape[0]) * (np.dot(dataset.T, errores))


for epoca in range(epocas):
 #Generar la predicción actual.
 predicciones = predecir(X, W, b)
 # calcular el error aislando la variable
 error = calcular_error(predicciones, y_real)
 # Monitorear el costo
 if epoca % 10 == 0:
      print(f"epocas {epoca} - Costo: {calcular_costo(error)}")
 # Calcular el gradiente de W (dw) y de b (db).
 dw = calculo_gradiente(X, error)
 db = (1 / X.shape[0]) * np.sum(error)
 W = W - alpha * dw
 b = b - alpha * db



# Datos
# Predicc
```

```python
import numpy as np



class RegresionLogistica:
  def __init__(self, alpha, epocas) -> None:
    self.__alpha = alpha
    self.__epocas = epocas
    self.__W = None
    self.__b = None

  def predict(self, X):
    z = np.dot(X, self.__W) + self.__b
    prediccion = self.__sigmoide(z)
    filtro = (prediccion >= 0.5).astype(int)
    return filtro

  def __sigmoide(self, prediccion: np.ndarray):
  # 1 / 1 + e ^ -z
    return (1/ (1 + np.exp(-prediccion)))

  def calcular_costo(self, predicciones: np.ndarray, real: np.ndarray) -> float:
    # -[real .log(prediccion) + (1 - real) . log(1-prediccion)]
    epsilon = 1e-15
    predicciones_seguras = np.clip(predicciones, epsilon, 1 - epsilon)
    costo = np.mean(-(real * np.log(predicciones_seguras) + (1 - real) * np.log(1 - predicciones_seguras)))
    return costo

  def fit(self, X, y) -> None:

    # INICIALIZACIÓN DINÁMICA (El modelo despierta)
    # Extraemos n (cantidad de columnas/características de X)(shape retorna un tupla (columnas/filas))
    n_caracteristicas = X.shape[1]
    # Creamos una matriz de ceros de tamaño (n, 1) e inicializamos b para no tener errores en la reasignacion (calculo final)
    self.__W = np.zeros((n_caracteristicas, 1))
    self.__b = 0.0

    for epoca in range(self.__epocas):

      predicciones = self.predict(X)

      error = predicciones - y

      if epoca % 10 == 0:
        print(f"epocas {epoca} - Costo: {self.calcular_costo(predicciones, y)}")

      dw = (1/X.shape[0]) * (np.dot(X.T, error))

      db = (1 / X.shape[0]) * np.sum(error)

      self.__W = self.__W - self.__alpha * dw
      self.__b = self.__b - self.__alpha * db



# Dataset: [Sueldo, Deudas] (Normalizados para simplificar)
X = np.array([
    [0.8, 0.1],  # Alto sueldo, baja deuda -> Debería ser 1 (Aprobado)
    [0.4, 0.8],  # Sueldo medio, alta deuda -> Debería ser 0 (Rechazado)
    [0.9, 0.2],  # Alto sueldo, baja deuda -> Debería ser 1
    [0.2, 0.9]   # Bajo sueldo, alta deuda -> Debería ser 0
])
y_real = np.array([[1], [0], [1], [0]])

x_nuevo = np.array([
    [0.7, 0.3],  # Cliente A
    [0.3, 0.7],  # Cliente B
    [0.9, 0.9]   # Cliente C
])

modelo = RegresionLogistica(alpha=0.00001, epocas=100)
modelo.fit(X, y_real)
predicciones_nuevas = modelo.predict(x_nuevo)


print(predicciones_nuevas)

```
