```python
from sklearn.linear_model import LogisticRegression
import numpy as np

# Datos Históricos [Estrés, Horas Extra, Salario ($K)]
X_empleados = np.array([
    [8.0, 15.0, 60.0],  # Alto estrés, muchas horas, salario bajo -> Renunció
    [3.0, 2.0, 120.0],  # Bajo estrés, pocas horas, salario alto -> Se quedó
    [7.5, 12.0, 65.0],  # ... -> Renunció
    [4.0, 5.0, 95.0],   # ... -> Se quedó
    [9.0, 20.0, 70.0],  # ... -> Renunció
    [2.0, 0.0, 110.0]   # ... -> Se quedó
])
y_churn = np.array([1, 0, 1, 0, 1, 0])

# Tu equipo actual de desarrolladores (Necesitamos saber quién está en riesgo)
equipo_actual = np.array([
    [8.5, 10.0, 55.0], # Dev A
    [3.5, 4.0, 105.0], # Dev B
    [7.0, 5.0, 115.0]  # Dev C (Alto estrés, pero salario muy alto y horas normales)
])

clf = LogisticRegression().fit(X_empleados,y_churn )
print("predict probabilidades")
print(clf.predict_proba(equipo_actual))
print("predict clasificatiorio")
print(clf.predict(equipo_actual))
```
